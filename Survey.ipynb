```python

# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import sem


survey_file = "WhereYouAt_Survey.csv"
```

```python
survey_results = pd.read_csv(survey_file, engine = 'python')
```

```python
survey_results
```

```python
outdoor_group=survey_results.groupby("How far are you willing to travel for outdoor exercise (hiking, biking, etc.)?")
How_Often_Essentials_group=survey_results.groupby("How often do you go on essential trips (to the grocery store, medicine, work (if you need to go to work)) since COVID (February 2020)?")
Time_Leaving_home_group=survey_results.groupby("When are you comfortable leaving your home?")
shelter_in_place_YN_group=survey_results.groupby("Has your attitude about shelter-in-place/stay-at-home/quarantine orders changed?")
```

```python

#plot df using pandas
outdoor_group.plot(kind="bar")
#title
plt.title("group_outdoor")
plt.xlabel("distance in miles")
plt.ylabel("Total Count")
plt.show()
```

```python
#plot df using pandas
How_Often_Essentials_group.plot(kind="bar")
#title
plt.title("How Often for Essentials")
plt.xlabel("often go out")
plt.ylabel("Total Count")
plt.show()
```

```python

#plot df using pandas

#plt.pie(Time_Leaving_home_group ,labels="What Time Leaving home",autopct='%1.1f%%')
#plt.ylabel("Preference to leave home")
#plt.show()
#Time_Leaving_home_group
Timing_count = Time_Leaving_home_group['When are you comfortable leaving your home?'].value_counts()
Timing_count.plot(kind="pie",autopct='%1.1f%%')
plt.pie(Timing_count.values ,autopct='%1.1f%%')
plt.ylabel("Preference time to leave home")
plt.show()

```
